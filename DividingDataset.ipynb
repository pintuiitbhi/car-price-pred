```python
import pandas as pd
import numpy as np

df = pd.read_csv('OLX_Car_Data_CSV.csv',encoding= 'latin1')

# df['split'] = np.random.randn(df.shape[0], 1)
df['Duplicated']=df.duplicated()
```

```python
df.head()
```

```python
df1=df.loc[df['Duplicated'] == True]
print("Number of duplicates in the dataset = ",df1.shape[0])
```

So let's remove them first

```python
df2= df.loc[df['Duplicated'] == False]
df2.head()
```

```python
df2.shape
```

```python
df2=df2.drop(['Duplicated'], axis=1)
df2.head()
```

```python
msk = np.random.rand(len(df2)) <= 0.7

train = df2[msk]
test = df2[~msk]
```

```python
train.to_csv("train.tsv", sep='\t', encoding='utf-8',index=False)
test.to_csv("test.tsv", sep='\t', encoding='utf-8',index=False)
```
